## Check for Missing Values (df1)

### Subtask:
Identify and quantify missing values in `df1` using `df1.isnull().sum()` and visualize the missing data percentage per column for better insight.


```python
import matplotlib.pyplot as plt

# Calculate the number of missing values for each column in df1
missing_values = df1.isnull().sum()
print("Number of Missing Values in df1:\n", missing_values)

# Calculate the percentage of missing values for each column in df1
missing_percentage = (df1.isnull().sum() / len(df1)) * 100
print("\nPercentage of Missing Values in df1:\n", missing_percentage)

# Filter out columns with no missing values for plotting
missing_percentage_to_plot = missing_percentage[missing_percentage > 0]

# Create a bar plot to visualize the percentage of missing values per column
plt.figure(figsize=(10, 6))
missing_percentage_to_plot.plot(kind='bar', color='skyblue')
plt.title('Percentage of Missing Values in df1')
plt.xlabel('Columns')
plt.ylabel('Percentage of Missing Values (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Visualize Key Distributions (df1)

### Subtask:
Create histograms and box plots for key numerical features in `df1` to visualize their distributions, identify potential outliers, and understand their spread. Ensure plots have appropriate titles and legends.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the numerical columns for visualization
numerical_cols = [
    'Level_of_Hemoglobin',
    'Genetic_Pedigree_Coefficient',
    'Age',
    'BMI',
    'salt_content_in_the_diet',
    'alcohol_consumption_per_day',
    'Level_of_Stress'
]

# Loop through each numerical column to create histograms and box plots
for col in numerical_cols:
    plt.figure(figsize=(12, 5))

    # Subplot 1: Histogram
    plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
    sns.histplot(df1[col].dropna(), kde=True, bins=30)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Subplot 2: Box Plot
    plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
    sns.boxplot(y=df1[col].dropna())
    plt.title(f'{col} Box Plot')
    plt.ylabel(col)

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np

# Create a mask for non-zero alcohol consumption
non_zero_alcohol = df1[df1['alcohol_consumption_per_day'] > 0]['alcohol_consumption_per_day']

# Calculate quartiles for non-zero alcohol consumption
# Using qcut with duplicates='drop' to handle cases where quartile boundaries might be the same
quartile_labels = pd.qcut(
    non_zero_alcohol,
    q=[0, 0.25, 0.5, 0.75, 1],
    labels=['Low', 'Medium', 'High', 'Very High'],
    duplicates='drop'
)

# Create the new binned column, starting with 'No Consumption'
df1['alcohol_consumption_quartile_category'] = 'No Consumption'

# Assign the quartile labels to the rows where alcohol_consumption_per_day > 0
df1.loc[df1['alcohol_consumption_per_day'] > 0, 'alcohol_consumption_quartile_category'] = quartile_labels

print("Value counts for 'alcohol_consumption_quartile_category':")
display(df1['alcohol_consumption_quartile_category'].value_counts(dropna=False))
```
